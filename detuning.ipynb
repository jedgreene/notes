# Mass-Spring System with Detuning in the Rotating Frame

Let’s work through the mass-spring system example with a driving force at a frequency close to the system's natural frequency. We’ll incorporate detuning and transform to the rotating frame.

### Problem Setup: Mass-Spring System with Detuning

We have a mass $m$ on a spring with:
- A natural frequency $\omega$,
- An external driving force oscillating at frequency $\omega_d$, which may differ slightly from $\omega$.

The detuning $\Delta$ is defined as:
$$
\Delta = \omega - \omega_d.
$$

The Hamiltonian of the system in the lab frame is:
$$
H = \frac{p^2}{2m} + \frac{1}{2} m \omega^2 x^2 + F_0 x \cos(\omega_d t),
$$
where:
- $\frac{p^2}{2m}$ is the kinetic energy,
- $\frac{1}{2} m \omega^2 x^2$ is the potential energy of the spring,
- $F_0 x \cos(\omega_d t)$ is the driving term, with $F_0$ being the driving force amplitude.

### Step 1: Moving to the Rotating Frame

To analyze the system in a rotating frame oscillating at the drive frequency $\omega_d$, we express the position $x$ in terms of a new, slowly varying variable $X$ in the rotating frame:
$$
x = X \cos(\omega_d t).
$$
This substitution allows us to “follow” the oscillations at $\omega_d$, simplifying our analysis of the system’s slow dynamics.

### Step 2: Substitute $x = X \cos(\omega_d t)$ into the Hamiltonian

We substitute $x = X \cos(\omega_d t)$ into each term of the Hamiltonian.

#### Kinetic Energy Term

The velocity $\dot{x}$ is:
$$
\dot{x} = \frac{d}{dt} (X \cos(\omega_d t)) = \dot{X} \cos(\omega_d t) - X \omega_d \sin(\omega_d t).
$$
The kinetic energy $\frac{p^2}{2m}$ becomes:
$$
\frac{p^2}{2m} = \frac{1}{2m} \left( m \dot{X} \cos(\omega_d t) - m X \omega_d \sin(\omega_d t) \right)^2.
$$
Expanding this, we get:
$$
\frac{p^2}{2m} = \frac{1}{2} m \dot{X}^2 \cos^2(\omega_d t) - m \dot{X} X \omega_d \cos(\omega_d t) \sin(\omega_d t) + \frac{1}{2} m X^2 \omega_d^2 \sin^2(\omega_d t).
$$
We now time-average each term over a period of the fast oscillations at $\omega_d$. Using:
1. $\langle \cos^2(\omega_d t) \rangle = \frac{1}{2}$,
2. $\langle \sin^2(\omega_d t) \rangle = \frac{1}{2}$,
3. $\langle \cos(\omega_d t) \sin(\omega_d t) \rangle = 0$,

the kinetic energy term becomes:
$$
\frac{p^2}{2m} \approx \frac{1}{4} m \dot{X}^2 + \frac{1}{4} m X^2 \omega_d^2.
$$

#### Potential Energy Term

For the potential energy $\frac{1}{2} m \omega^2 x^2$, we substitute $x = X \cos(\omega_d t)$:
$$
\frac{1}{2} m \omega^2 x^2 = \frac{1}{2} m \omega^2 X^2 \cos^2(\omega_d t).
$$
Averaging over time, this becomes:
$$
\frac{1}{4} m \omega^2 X^2.
$$

#### Driving Force Term

The driving term $F_0 x \cos(\omega_d t)$ becomes:
$$
F_0 x \cos(\omega_d t) = F_0 X \cos(\omega_d t) \cos(\omega_d t) = F_0 X \cos^2(\omega_d t).
$$
Averaging $\cos^2(\omega_d t)$ over time gives:
$$
\frac{1}{2} F_0 X.
$$

### Step 3: Introducing the Detuning Term

Now, let’s look closely at the combined potential-like terms from the kinetic and potential energies:
1. From the kinetic energy term, we have $\frac{1}{4} m X^2 \omega_d^2$.
2. From the potential energy term, we have $\frac{1}{4} m X^2 \omega^2$.

To incorporate detuning, we expand $\omega^2$ in terms of $\omega_d$ and $\Delta = \omega - \omega_d$:
$$
\omega^2 = (\omega_d + \Delta)^2 = \omega_d^2 + 2 \omega_d \Delta + \Delta^2.
$$
Substituting this expansion, the potential energy term becomes:
$$
\frac{1}{4} m X^2 \omega^2 = \frac{1}{4} m X^2 (\omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
Now we combine this with the kinetic contribution $\frac{1}{4} m X^2 \omega_d^2$:
$$
\frac{1}{4} m X^2 \omega_d^2 + \frac{1}{4} m X^2 (\omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
This simplifies to:
$$
\frac{1}{4} m X^2 (2 \omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
For small detuning (where $\Delta$ is much smaller than $\omega_d$), we can neglect the $\Delta^2$ term, giving:
$$
\frac{1}{2} m X^2 \omega_d^2 + \frac{1}{2} m X^2 \omega_d \Delta.
$$

### Final Rotating Frame Hamiltonian with Detuning

The Hamiltonian in the rotating frame, including the detuning term, is:
$$
H_R = \frac{P^2}{2m} + \frac{1}{2} m X^2 \omega_d^2 + \frac{1}{2} m X^2 \omega_d \Delta + \frac{1}{2} F_0 X.
$$
This Hamiltonian includes the detuning term:
$$
\frac{1}{2} m X^2 \omega_d \Delta,
$$
which reflects the frequency mismatch in the effective potential in the rotating frame.

### General Form of the Detuning Term

This specific example confirms the general form for incorporating detuning in the rotating frame:
$$
\text{Detuning term} = \frac{1}{2} m X^2 \omega_d \Delta.
$$
In general, when we transform to a rotating frame at the drive frequency $\omega_d$, we can add this term directly to account for any detuning $\Delta = \omega - \omega_d$, avoiding the need for a full derivation each time.
