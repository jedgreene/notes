# Mass-Spring System with Detuning in the Rotating Frame

Let’s work through the mass-spring system example with a driving force at a frequency close to the system's natural frequency. We’ll incorporate detuning and transform to the rotating frame.

### Problem Setup: Mass-Spring System with Detuning

We have a mass $m$ on a spring with:
- A natural frequency $\omega$,
- An external driving force oscillating at frequency $\omega_d$, which may differ slightly from $\omega$.

The detuning $\Delta$ is defined as:
$$
\Delta = \omega - \omega_d.
$$

The Hamiltonian of the system in the lab frame is:
$$
H = \frac{p^2}{2m} + \frac{1}{2} m \omega^2 x^2 + F_0 x \cos(\omega_d t),
$$
where:
- $\frac{p^2}{2m}$ is the kinetic energy,
- $\frac{1}{2} m \omega^2 x^2$ is the potential energy of the spring,
- $F_0 x \cos(\omega_d t)$ is the driving term, with $F_0$ being the driving force amplitude.

### Step 1: Moving to the Rotating Frame

To analyze the system in a rotating frame oscillating at the drive frequency $\omega_d$, we express the position $x$ in terms of a new, slowly varying variable $X$ in the rotating frame:
$$
x = X \cos(\omega_d t).
$$
This substitution allows us to “follow” the oscillations at $\omega_d$, simplifying our analysis of the system’s slow dynamics.

### Step 2: Substitute $x = X \cos(\omega_d t)$ into the Hamiltonian

We substitute $x = X \cos(\omega_d t)$ into each term of the Hamiltonian.

#### Kinetic Energy Term

The velocity $\dot{x}$ is:
$$
\dot{x} = \frac{d}{dt} (X \cos(\omega_d t)) = \dot{X} \cos(\omega_d t) - X \omega_d \sin(\omega_d t).
$$
The kinetic energy $\frac{p^2}{2m}$ becomes:
$$
\frac{p^2}{2m} = \frac{1}{2m} \left( m \dot{X} \cos(\omega_d t) - m X \omega_d \sin(\omega_d t) \right)^2.
$$
Expanding this, we get:
$$
\frac{p^2}{2m} = \frac{1}{2} m \dot{X}^2 \cos^2(\omega_d t) - m \dot{X} X \omega_d \cos(\omega_d t) \sin(\omega_d t) + \frac{1}{2} m X^2 \omega_d^2 \sin^2(\omega_d t).
$$
We now time-average each term over a period of the fast oscillations at $\omega_d$. Using:
1. $\langle \cos^2(\omega_d t) \rangle = \frac{1}{2}$,
2. $\langle \sin^2(\omega_d t) \rangle = \frac{1}{2}$,
3. $\langle \cos(\omega_d t) \sin(\omega_d t) \rangle = 0$,

the kinetic energy term becomes:
$$
\frac{p^2}{2m} \approx \frac{1}{4} m \dot{X}^2 + \frac{1}{4} m X^2 \omega_d^2.
$$

#### Potential Energy Term

For the potential energy $\frac{1}{2} m \omega^2 x^2$, we substitute $x = X \cos(\omega_d t)$:
$$
\frac{1}{2} m \omega^2 x^2 = \frac{1}{2} m \omega^2 X^2 \cos^2(\omega_d t).
$$
Averaging over time, this becomes:
$$
\frac{1}{4} m \omega^2 X^2.
$$

#### Driving Force Term

The driving term $F_0 x \cos(\omega_d t)$ becomes:
$$
F_0 x \cos(\omega_d t) = F_0 X \cos(\omega_d t) \cos(\omega_d t) = F_0 X \cos^2(\omega_d t).
$$
Averaging $\cos^2(\omega_d t)$ over time gives:
$$
\frac{1}{2} F_0 X.
$$

### Step 3: Introducing the Detuning Term

Now, let’s look closely at the combined potential-like terms from the kinetic and potential energies:
1. From the kinetic energy term, we have $\frac{1}{4} m X^2 \omega_d^2$.
2. From the potential energy term, we have $\frac{1}{4} m X^2 \omega^2$.

To incorporate detuning, we expand $\omega^2$ in terms of $\omega_d$ and $\Delta = \omega - \omega_d$:
$$
\omega^2 = (\omega_d + \Delta)^2 = \omega_d^2 + 2 \omega_d \Delta + \Delta^2.
$$
Substituting this expansion, the potential energy term becomes:
$$
\frac{1}{4} m X^2 \omega^2 = \frac{1}{4} m X^2 (\omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
Now we combine this with the kinetic contribution $\frac{1}{4} m X^2 \omega_d^2$:
$$
\frac{1}{4} m X^2 \omega_d^2 + \frac{1}{4} m X^2 (\omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
This simplifies to:
$$
\frac{1}{4} m X^2 (2 \omega_d^2 + 2 \omega_d \Delta + \Delta^2).
$$
For small detuning (where $\Delta$ is much smaller than $\omega_d$), we can neglect the $\Delta^2$ term, giving:
$$
\frac{1}{2} m X^2 \omega_d^2 + \frac{1}{2} m X^2 \omega_d \Delta.
$$

### Final Rotating Frame Hamiltonian with Detuning

The Hamiltonian in the rotating frame, including the detuning term, is:
$$
H_R = \frac{P^2}{2m} + \frac{1}{2} m X^2 \omega_d^2 + \frac{1}{2} m X^2 \omega_d \Delta + \frac{1}{2} F_0 X.
$$
This Hamiltonian includes the detuning term:
$$
\frac{1}{2} m X^2 \omega_d \Delta,
$$
which reflects the frequency mismatch in the effective potential in the rotating frame.

### General Form of the Detuning Term

This specific example confirms the general form for incorporating detuning in the rotating frame:
$$
\text{Detuning term} = \frac{1}{2} m X^2 \omega_d \Delta.
$$
In general, when we transform to a rotating frame at the drive frequency $\omega_d$, we can add this term directly to account for any detuning $\Delta = \omega - \omega_d$, avoiding the need for a full derivation each time.


# Deriving the Rotating Frame Hamiltonian $H_{\text{rot}}$

Starting from our lab frame Hamiltonian:
$$
H = \frac{\hbar \omega_0}{2} \sigma_z + \hbar \Omega \cos(\omega_d t) \sigma_x
$$

we apply the unitary transformation $U(t) = e^{i \frac{\omega_d}{2} \sigma_z t}$ to move to a frame rotating at the driving frequency $\omega_d$. In this rotating frame, the transformed state vector is:
$$
|\psi_{\text{rot}}(t)\rangle = U(t) |\psi(t)\rangle
$$

We now find the Hamiltonian $H_{\text{rot}}$ such that $|\psi_{\text{rot}}(t)\rangle$ evolves according to:
$$
i \hbar \frac{d}{dt} |\psi_{\text{rot}}(t)\rangle = H_{\text{rot}} |\psi_{\text{rot}}(t)\rangle
$$

### Step-by-Step Derivation

1. **Express $|\psi(t)\rangle$ in Terms of $|\psi_{\text{rot}}(t)\rangle$:**  
   Since $|\psi(t)\rangle = U^\dagger(t) |\psi_{\text{rot}}(t)\rangle$, differentiating with respect to $t$ gives:
   $$
   \frac{d}{dt} |\psi(t)\rangle = \frac{d}{dt} \left( U^\dagger(t) |\psi_{\text{rot}}(t)\rangle \right)
   $$
   Applying the product rule:
   $$
   \frac{d}{dt} |\psi(t)\rangle = \left( \frac{d}{dt} U^\dagger(t) \right) |\psi_{\text{rot}}(t)\rangle + U^\dagger(t) \frac{d}{dt} |\psi_{\text{rot}}(t)\rangle
   $$

2. **Rewrite the Schrödinger Equation:**  
   In the lab frame, the Schrödinger equation is:
   $$
   i \hbar \frac{d}{dt} |\psi(t)\rangle = H |\psi(t)\rangle
   $$
   Substituting the expression for $\frac{d}{dt} |\psi(t)\rangle$ from above, we get:
   $$
   i \hbar \left( \left( \frac{d}{dt} U^\dagger(t) \right) |\psi_{\text{rot}}(t)\rangle + U^\dagger(t) \frac{d}{dt} |\psi_{\text{rot}}(t)\rangle \right) = H U^\dagger(t) |\psi_{\text{rot}}(t)\rangle
   $$

3. **Multiply Both Sides by $U(t)$:**  
   To simplify, multiply both sides by $U(t)$:
   $$
   i \hbar U(t) \frac{d}{dt} U^\dagger(t) |\psi_{\text{rot}}(t)\rangle + i \hbar \frac{d}{dt} |\psi_{\text{rot}}(t)\rangle = U(t) H U^\dagger(t) |\psi_{\text{rot}}(t)\rangle
   $$

4. **Isolate $\frac{d}{dt} |\psi_{\text{rot}}(t)\rangle$:**  
   Rearrange terms to isolate $\frac{d}{dt} |\psi_{\text{rot}}(t)\rangle$:
   $$
   i \hbar \frac{d}{dt} |\psi_{\text{rot}}(t)\rangle = \left( U(t) H U^\dagger - i \hbar U(t) \frac{d}{dt} U^\dagger(t) \right) |\psi_{\text{rot}}(t)\rangle
   $$

Thus, the effective Hamiltonian in the rotating frame is:
$$
H_{\text{rot}} = U H U^\dagger - i \hbar U \frac{d}{dt} U^\dagger
$$

This equation shows how the Hamiltonian $H$ transforms under a time-dependent unitary transformation $U(t)$. The resulting $H_{\text{rot}}$ describes the system’s dynamics in the rotating frame.


## Derivation of the Detuning Term with Detailed Application of RWA

We start with the Hamiltonian for a two-level system with a driving field:

$$
H = \frac{\hbar \omega_0}{2} \sigma_z + \hbar \Omega \cos(\omega t) \sigma_x
$$

where:
- $\omega_0$ is the natural frequency of the system,
- $\Omega$ is the amplitude of the driving field (Rabi frequency),
- $\omega$ is the frequency of the driving field.

### Step 1: Express the Driving Term Using Complex Exponentials

Rewrite $\cos(\omega t)$ as:

$$
\cos(\omega t) = \frac{1}{2} \left( e^{i \omega t} + e^{-i \omega t} \right)
$$

so that the driving Hamiltonian becomes:

$$
H_{\text{drive}} = \frac{\hbar \Omega}{2} \left( e^{i \omega t} + e^{-i \omega t} \right) \sigma_x
$$

Then, the total Hamiltonian is:

$$
H = \frac{\hbar \omega_0}{2} \sigma_z + \frac{\hbar \Omega}{2} \left( e^{i \omega t} + e^{-i \omega t} \right) \sigma_x
$$

### Step 2: Moving to the Rotating Frame

To simplify further, we move to a rotating frame at frequency $\omega$ using:

$$
U(t) = e^{i \frac{\omega}{2} t \sigma_z}
$$

The Hamiltonian in the rotating frame is then given by:

$$
H' = U H U^\dagger - i \hbar U \frac{d}{dt} U^\dagger
$$

### Computing the Time Derivative Term

To calculate the time derivative term, we first compute the derivative of $U(t)$ with respect to time:

1. Start with \( U(t) = e^{i \frac{\omega}{2} t \sigma_z} \).

2. Differentiate \( U(t) \) with respect to \( t \):

   $$
   \frac{d}{dt} U(t) = i \frac{\omega}{2} \sigma_z e^{i \frac{\omega}{2} t \sigma_z} = i \frac{\omega}{2} \sigma_z U(t)
   $$

3. Now, calculate \( U \frac{d}{dt} U^\dagger \):

   Since \( U^\dagger = e^{-i \frac{\omega}{2} t \sigma_z} \), we get:

   $$
   U \frac{d}{dt} U^\dagger = U \left( i \frac{\omega}{2} \sigma_z \right) U^\dagger
   $$

4. Using the fact that \( U \sigma_z U^\dagger = \sigma_z \) (since \(\sigma_z\) commutes with itself), we find:

   $$
   U \frac{d}{dt} U^\dagger = i \frac{\omega}{2} \sigma_z
   $$

5. Therefore, the time derivative term is:

   $$
   -i \hbar U \frac{d}{dt} U^\dagger = -i \hbar \left( i \frac{\omega}{2} \sigma_z \right) = \frac{\hbar \omega}{2} \sigma_z
   $$

### Step 3: Transform Each Term in the Hamiltonian

#### Transforming $ H_0 = \frac{\hbar \omega_0}{2} \sigma_z $

Since $\sigma_z$ commutes with itself, this term remains the same:

$$
U H_0 U^\dagger = \frac{\hbar \omega_0}{2} \sigma_z
$$

#### Transforming $ H_{\text{drive}} = \frac{\hbar \Omega}{2} \left( e^{i \omega t} + e^{-i \omega t} \right) \sigma_x $

Using the identity $ e^{i \frac{\omega}{2} t \sigma_z} \sigma_x e^{-i \frac{\omega}{2} t \sigma_z} = \cos(\omega t) \sigma_x + \sin(\omega t) \sigma_y $, we get:

$$
U H_{\text{drive}} U^\dagger = \frac{\hbar \Omega}{2} \left( e^{i \omega t} (\cos(\omega t) \sigma_x + \sin(\omega t) \sigma_y) + e^{-i \omega t} (\cos(\omega t) \sigma_x + \sin(\omega t) \sigma_y) \right)
$$

Expanding each product, we obtain four terms:
1. $ \frac{\hbar \Omega}{2} e^{i \omega t} \cos(\omega t) \sigma_x $
2. $ \frac{\hbar \Omega}{2} e^{i \omega t} \sin(\omega t) \sigma_y $
3. $ \frac{\hbar \Omega}{2} e^{-i \omega t} \cos(\omega t) \sigma_x $
4. $ \frac{\hbar \Omega}{2} e^{-i \omega t} \sin(\omega t) \sigma_y $

### Step 4: Expanding and Applying the Rotating Wave Approximation (RWA)

We now expand each term using $\cos(\omega t) = \frac{1}{2} (e^{i \omega t} + e^{-i \omega t})$ and $\sin(\omega t) = \frac{1}{2i} (e^{i \omega t} - e^{-i \omega t})$, and apply the RWA by discarding terms oscillating at high frequencies (such as $e^{\pm 2 i \omega t}$), as they average out over time.

- **First term:** $ \frac{\hbar \Omega}{2} e^{i \omega t} \cos(\omega t) \sigma_x $
  
  Expanding $\cos(\omega t)$:

  $$
  \frac{\hbar \Omega}{2} e^{i \omega t} \cos(\omega t) \sigma_x = \frac{\hbar \Omega}{4} \left( e^{2i \omega t} + 1 \right) \sigma_x
  $$

  Applying the RWA, we discard the rapidly oscillating term $e^{2i \omega t}$, leaving:

  $$
  \frac{\hbar \Omega}{4} \sigma_x
  $$

- **Second term:** $ \frac{\hbar \Omega}{2} e^{i \omega t} \sin(\omega t) \sigma_y $

  Expanding $\sin(\omega t)$:

  $$
  \frac{\hbar \Omega}{2} e^{i \omega t} \sin(\omega t) \sigma_y = \frac{\hbar \Omega}{4i} \left( e^{2i \omega t} - 1 \right) \sigma_y
  $$

  Applying the RWA and discarding $e^{2i \omega t}$, we are left with:

  $$
  -\frac{\hbar \Omega}{4i} \sigma_y
  $$

- **Third term:** $ \frac{\hbar \Omega}{2} e^{-i \omega t} \cos(\omega t) \sigma_x $

  Similarly, expanding $\cos(\omega t)$:

  $$
  \frac{\hbar \Omega}{2} e^{-i \omega t} \cos(\omega t) \sigma_x = \frac{\hbar \Omega}{4} \left( 1 + e^{-2i \omega t} \right) \sigma_x
  $$

  Applying the RWA (discarding $e^{-2i \omega t}$), we get:

  $$
  \frac{\hbar \Omega}{4} \sigma_x
  $$

- **Fourth term:** $ \frac{\hbar \Omega}{2} e^{-i \omega t} \sin(\omega t) \sigma_y $

  Expanding $\sin(\omega t)$:

  $$
  \frac{\hbar \Omega}{2} e^{-i \omega t} \sin(\omega t) \sigma_y = \frac{\hbar \Omega}{4i} \left( 1 - e^{-2i \omega t} \right) \sigma_y
  $$

  Applying the RWA (discarding $e^{-2i \omega t}$), we are left with:

  $$
  \frac{\hbar \Omega}{4i} \sigma_y
  $$

### Combining Terms After RWA

After applying the RWA, the contributions from all four terms simplify to:

$$
H'_{\text{drive}} \approx \frac{\hbar \Omega}{2} \sigma_x
$$

#### Time Derivative Term

The time derivative term from the rotating frame transformation is:

$$
-i \hbar U \frac{d}{dt} U^\dagger = \frac{\hbar \omega}{2} \sigma_z
$$

### Final Rotating Frame Hamiltonian with Detuning

Combining all contributions, the Hamiltonian in the rotating frame becomes:

$$
H' = \frac{\hbar \omega_0}{2} \sigma_z + \frac{\hbar \Omega}{2} \sigma_x - \frac{\hbar \omega}{2} \sigma_z
$$

Combining $\sigma_z$ terms, this simplifies to:

$$
H' = \frac{\hbar (\omega_0 - \omega)}{2} \sigma_z + \frac{\hbar \Omega}{2} \sigma_x
$$

### Introducing the Detuning $\Delta$

Define the **detuning** $\Delta$ as the difference between the system frequency $\omega_0$ and the driving frequency $\omega$:

$$
\Delta = \omega - \omega_0
$$

Thus, the rotating frame Hamiltonian becomes:

$$
H' = \frac{\hbar \Delta}{2} \sigma_z + \frac{\hbar \Omega}{2} \sigma_x
$$

### Final Result

The term $\frac{\hbar \Delta}{2} \sigma_z$ represents the **detuning**, which shows how the difference between the system’s natural frequency and the drive frequency affects the system's dynamics. When $\Delta = 0$, the drive is on resonance, leading to maximum state transitions; when $\Delta \neq 0$, the system is off-resonant, modifying oscillations and enabling selective control over the state populations.
