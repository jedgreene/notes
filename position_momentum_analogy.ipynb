## Simplified Harmonic Oscillator Hamiltonian

To see if treating **phase** or **charge** as the “position-like” variable leads to identical results, let’s consider a simplified harmonic oscillator model by ignoring the nonlinearity entirely.

### Hamiltonian

Starting from the transmon-like Hamiltonian, we ignore the nonlinear (quartic) term and focus only on the harmonic oscillator part:

$$
\hat{H} \approx \frac{\hat{Q}^2}{2C} + \frac{\hat{\Phi}^2}{2L_J}
$$

where:
- $\hat{Q}$ is the charge operator, representing the charge across the junction,
- $C$ is the capacitance,
- $\hat{\Phi}$ is the phase operator, representing the phase difference across the junction,
- $L_J$ is the effective inductance, defined as $L_J = \frac{\phi_0^2}{E_J}$.

This Hamiltonian resembles a simple harmonic oscillator with effective capacitance $C$ and inductance $L_J$.

### Case A: Phase as the "Position-Like" Variable

1. **Assumptions**:
   - Assume that $\hat{\Phi}$ is the "position-like" variable.
   - Since $\hat{\Phi}$ and $\hat{Q}$ are conjugate variables, $\hat{Q}$ will play the role of "momentum."

2. **Representation in Terms of Creation and Annihilation Operators**:
   - We represent the operators in terms of creation and annihilation operators, assuming:
     $$
     \hat{\Phi} = \Phi_{ZPF} (\hat{a} + \hat{a}^\dagger), \quad \hat{Q} = i Q_{ZPF} (\hat{a}^\dagger - \hat{a})
     $$
   - Here, $\Phi_{ZPF} = \sqrt{\frac{\hbar Z}{2}}$ and $Q_{ZPF} = \sqrt{\frac{\hbar}{2Z}}$ represent the zero-point fluctuation amplitudes of phase and charge, respectively, with $Z = \sqrt{L_J / C}$.

3. **Resulting Hamiltonian**:
   - Substituting these into the Hamiltonian gives:
     $$
     \hat{H} = \hbar \omega_0 \left( \hat{a}^\dagger \hat{a} + \frac{1}{2} \right)
     $$
   - where $\omega_0 = \frac{1}{\sqrt{L_J C}}$ is the natural frequency of the harmonic oscillator.
   - The quantized energy levels are:
     $$
     E_n = \hbar \omega_0 \left( n + \frac{1}{2} \right)
     $$

### Case B: Charge as the "Position-Like" Variable

1. **Assumptions**:
   - Assume that $\hat{Q}$ is the "position-like" variable.
   - In this interpretation, $\hat{\Phi}$ will play the role of "momentum."

2. **Representation in Terms of Creation and Annihilation Operators**:
   - We write:
     $$
     \hat{Q} = Q_{ZPF} (\hat{a} + \hat{a}^\dagger), \quad \hat{\Phi} = i \Phi_{ZPF} (\hat{a}^\dagger - \hat{a})
     $$

3. **Resulting Hamiltonian**:
   - Substituting these into the Hamiltonian gives:
     $$
     \hat{H} = \hbar \omega_0 \left( \hat{a}^\dagger \hat{a} + \frac{1}{2} \right)
     $$
   - with $\omega_0 = \frac{1}{\sqrt{L_J C}}$ as before.
   - The quantized energy levels are:
     $$
     E_n = \hbar \omega_0 \left( n + \frac{1}{2} \right)
     $$

### Conclusion: Identical Results in Both Views

In this purely harmonic case, whether we treat **phase** $\hat{\Phi}$ or **charge** $\hat{Q}$ as the "position-like" variable, we obtain the **same Hamiltonian and identical quantized energy levels**:

$$
E_n = \hbar \omega_0 \left( n + \frac{1}{2} \right)
$$

This confirms that in a purely quadratic (harmonic) system, the choice of "position-like" variable is arbitrary, as long as we treat the conjugate variable as "momentum." The harmonic oscillator’s symmetric structure ensures identical results under either interpretation.
