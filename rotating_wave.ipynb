### Original Hamiltonian
The initial Hamiltonian for a qubit without any external drive is:
$$
H = \frac{\hbar \omega_0}{2} \sigma_z,
$$
where $\omega_0$ is the natural frequency of the qubit.

### Rotating Frame Transformation
The unitary operator for moving to a rotating frame at frequency $\omega$ is:
$$
U(t) = e^{i \frac{\omega t}{2} \sigma_z}.
$$

### Rotating Frame Hamiltonian
The Hamiltonian in the rotating frame is:
$$
H_{\text{rot}} = U H U^\dagger - i\hbar U \frac{dU^\dagger}{dt}.
$$

### Step-by-Step Derivation

1. **Compute $U \frac{dU^\dagger}{dt}$:**
   $$
   U^\dagger = e^{-i \frac{\omega t}{2} \sigma_z},
   $$
   $$
   \frac{dU^\dagger}{dt} = -i \frac{\omega}{2} \sigma_z U^\dagger,
   $$
   $$
   U \frac{dU^\dagger}{dt} = -i \frac{\omega}{2} \sigma_z.
   $$
   Therefore:
   $$
   -i\hbar U \frac{dU^\dagger}{dt} = -\frac{\hbar \omega}{2} \sigma_z.
   $$

2. **Transform the Original Hamiltonian $H$:**
   $$
   U H U^\dagger = \frac{\hbar \omega_0}{2} \sigma_z.
   $$

3. **Substitute into the Rotating Frame Hamiltonian:**
   $$
   H_{\text{rot}} = \frac{\hbar \omega_0}{2} \sigma_z - \frac{\hbar \omega}{2} \sigma_z.
   $$

4. **Combine the Terms:**
   $$
   H_{\text{rot}} = \frac{\hbar (\omega_0 - \omega)}{2} \sigma_z.
   $$

### Final Result
The Hamiltonian in the rotating frame is:
$$
H_{\text{rot}} = \frac{\hbar (\omega_0 - \omega)}{2} \sigma_z.
$$

This result represents the detuning between the natural frequency of the qubit and the frequency of the rotating frame.
