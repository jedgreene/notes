# Circuit Quantum Electrodynamics (cQED)

In the context of a qubit coupled to a resonator, the exchange of virtual photons leads to state-dependent energy shifts in the system. This setup falls within the domain of **circuit quantum electrodynamics (cQED)**, which serves as a bridge between quantum mechanics and quantum field theory (QFT).

While QFT deals with relativistic quantum fields and encompasses the creation and annihilation of particles in continuous, infinite-dimensional spaces, cQED operates within the framework of non-relativistic quantum mechanics. In cQED, the electromagnetic field is confined within a cavity or resonator, and the field modes are quantized, treating them as harmonic oscillators. The qubit is modeled as a discrete two-level system, such as a transmon qubit.

The concept of **virtual photons** in cQED arises from the interaction between the qubit and the quantized field of the resonator. These virtual photons are manifestations of the non-energy-conserving processes accounted for in second-order perturbation theory. They are not real, detectable particles but represent intermediate states that contribute to observable phenomena like state-dependent energy shifts and the Lamb shift.

Although cQED borrows language and concepts from QFT, such as virtual particles and quantized fields, it remains rooted in quantum mechanics. It focuses on discrete energy levels and finite-dimensional Hilbert spaces, avoiding the complexities associated with relativistic effects and infinite degrees of freedom in QFT.

In summary, cQED provides a powerful framework for studying the interaction between light and matter at the quantum level within engineered systems. It allows us to explore quantum phenomena like virtual photon exchange and state-dependent energy shifts without invoking the full machinery of quantum field theory.


# State-Dependent Energy Shifts in a Transmon Qubit Coupled to a Resonator

When a transmon qubit is coupled to a resonator, **virtual photon interactions** lead to **state-dependent energy shifts** in the system. These shifts arise from second-order perturbation processes involving virtual transitions between the qubit and the resonator. This explanation will detail these energy shifts using second-order perturbation theory, focusing on two key virtual processes:

1. The qubit in the ground state $\lvert 0 \rangle$ virtually **absorbs** a photon from the resonator, transitioning to the excited state $\lvert 1 \rangle$.
2. The qubit in the excited state $\lvert 1 \rangle$ virtually **emits** a photon into the resonator, transitioning to the ground state $\lvert 0 \rangle$.

---

## 1. System Hamiltonian

### a. Unperturbed Hamiltonian $H_0$

The unperturbed Hamiltonian $H_0$ describes the qubit and the resonator without interaction:

$$
H_0 = \frac{1}{2} \hbar \omega_q\, \hat{\sigma}_z + \hbar \omega_r\, \hat{a}^\dagger \hat{a}
$$

- $\hbar$: Reduced Planck's constant.
- $\omega_q$: Qubit transition frequency.
- $\omega_r$: Resonator frequency.
- $\hat{\sigma}_z$: Pauli $z$-operator for the qubit ($\hat{\sigma}_z = \lvert 1 \rangle \langle 1 \rvert - \lvert 0 \rangle \langle 0 \rvert$).
- $\hat{a}^\dagger$, $\hat{a}$: Creation and annihilation operators for the resonator photons.

**Energy Eigenstates and Energies:**

- **Qubit States:**
  - Ground state $\lvert 0 \rangle$ with energy $E_0^{(0)} = -\frac{1}{2} \hbar \omega_q$.
  - Excited state $\lvert 1 \rangle$ with energy $E_1^{(0)} = +\frac{1}{2} \hbar \omega_q$.

- **Resonator States:**
  - Photon number states $\lvert n \rangle$ with energy $E_n^{(0)} = \hbar \omega_r n$.

*Note:* The superscript $(0)$ on the energies $E^{(0)}$ denotes **unperturbed energies**, calculated from $H_0$ before considering the interaction $V$.

### b. Interaction Hamiltonian $V$

The interaction Hamiltonian describes the coupling between the qubit and the resonator:

$$
V = \hbar g \left( \hat{\sigma}_+ \hat{a} + \hat{\sigma}_- \hat{a}^\dagger \right)
$$

- $g$: Coupling strength between the qubit and the resonator.
- $\hat{\sigma}_+ = \lvert 1 \rangle \langle 0 \rvert$: Qubit raising operator.
- $\hat{\sigma}_- = \lvert 0 \rangle \langle 1 \rvert$: Qubit lowering operator.

**Explanation:**

- $\hat{\sigma}_+ \hat{a}$: The qubit in $\lvert 0 \rangle$ **absorbs** a photon from the resonator, transitioning to $\lvert 1 \rangle$.
- $\hat{\sigma}_- \hat{a}^\dagger$: The qubit in $\lvert 1 \rangle$ **emits** a photon into the resonator, transitioning to $\lvert 0 \rangle$.

---

## 2. Second-Order Perturbation Theory

The energy shift $\Delta E_n$ for a non-degenerate state $\lvert n \rangle$ in second-order perturbation theory is:

$$
\Delta E_n = \sum_{k \ne n} \frac{ \lvert \langle k \lvert V \rvert n \rangle \rvert^2 }{ E_n^{(0)} - E_k^{(0)} }
$$

- $\lvert n \rangle$: Initial state ($\lvert q, n \rangle = \lvert q \rangle \otimes \lvert n \rangle$, where $q = 0$ or $1$).
- $\lvert k \rangle$: Intermediate (virtual) states.
- $E_n^{(0)}$, $E_k^{(0)}$: Unperturbed energies from $H_0$.

---

## 3. Case 1: Qubit in $\lvert 0 \rangle$ Virtually Absorbing a Photon

### a. Initial State $\lvert 0, n \rangle$

- Qubit in ground state $\lvert 0 \rangle$.
- Resonator with $n$ photons.

### b. Intermediate State $\lvert 1, n - 1 \rangle$

- Qubit transitions to the excited state $\lvert 1 \rangle$.
- Resonator loses one photon ($n \rightarrow n - 1$).

### c. Matrix Element Calculation

$$
\begin{align*}
\langle 1, n - 1 \lvert V \rvert 0, n \rangle &= \hbar g\, \langle 1 \lvert \hat{\sigma}_+ \rvert 0 \rangle\, \langle n - 1 \lvert \hat{a} \rvert n \rangle \\
&= \hbar g\, (1)\, \sqrt{ n } \\
&= \hbar g\, \sqrt{ n }
\end{align*}
$$

### d. Energy Denominator

$$
\begin{align*}
E_{0, n}^{(0)} - E_{1, n - 1}^{(0)} &= \left( -\frac{1}{2} \hbar \omega_q + \hbar \omega_r n \right) - \left( +\frac{1}{2} \hbar \omega_q + \hbar \omega_r ( n - 1 ) \right) \\
&= -\hbar \omega_q + \hbar \omega_r \\
&= -\hbar ( \omega_q - \omega_r ) \\
&= -\hbar \Delta
\end{align*}
$$

- $\Delta = \omega_q - \omega_r$: **Detuning** between the qubit and resonator frequencies.

### e. Second-Order Energy Shift

$$
\begin{align*}
\Delta E_{0, n} &= \frac{ \lvert \langle 1, n - 1 \lvert V \rvert 0, n \rangle \rvert^2 }{ E_{0, n}^{(0)} - E_{1, n - 1}^{(0)} } \\
&= \frac{ \left( \hbar g\, \sqrt{ n } \right)^2 }{ -\hbar \Delta } \\
&= -\frac{ \hbar g^2 n }{ \Delta }
\end{align*}
$$

**Interpretation:**

- The energy shift $\Delta E_{0, n}$ is **negative**, indicating a lowering of the energy of the state $\lvert 0, n \rangle$.
- This shift results from the qubit in $\lvert 0 \rangle$ virtually **absorbing** a photon from the resonator.

---

## 4. Case 2: Qubit in $\lvert 1 \rangle$ Virtually Emitting a Photon

### a. Initial State $\lvert 1, n \rangle$

- Qubit in the excited state $\lvert 1 \rangle$.
- Resonator with $n$ photons.

### b. Intermediate State $\lvert 0, n + 1 \rangle$

- Qubit transitions to the ground state $\lvert 0 \rangle$.
- Resonator gains one photon ($n \rightarrow n + 1$).

### c. Matrix Element Calculation

$$
\begin{align*}
\langle 0, n + 1 \lvert V \rvert 1, n \rangle &= \hbar g\, \langle 0 \lvert \hat{\sigma}_- \rvert 1 \rangle\, \langle n + 1 \lvert \hat{a}^\dagger \rvert n \rangle \\
&= \hbar g\, (1)\, \sqrt{ n + 1 } \\
&= \hbar g\, \sqrt{ n + 1 }
\end{align*}
$$

### d. Energy Denominator

$$
\begin{align*}
E_{1, n}^{(0)} - E_{0, n + 1}^{(0)} &= \left( +\frac{1}{2} \hbar \omega_q + \hbar \omega_r n \right) - \left( -\frac{1}{2} \hbar \omega_q + \hbar \omega_r ( n + 1 ) \right) \\
&= \hbar \omega_q - \hbar \omega_r \\
&= \hbar ( \omega_q - \omega_r ) \\
&= \hbar \Delta
\end{align*}
$$

### e. Second-Order Energy Shift

$$
\begin{align*}
\Delta E_{1, n} &= \frac{ \lvert \langle 0, n + 1 \lvert V \rvert 1, n \rangle \rvert^2 }{ E_{1, n}^{(0)} - E_{0, n + 1}^{(0)} } \\
&= \frac{ \left( \hbar g\, \sqrt{ n + 1 } \right)^2 }{ \hbar \Delta } \\
&= \frac{ \hbar g^2 ( n + 1 ) }{ \Delta }
\end{align*}
$$

**Interpretation:**

- The energy shift $\Delta E_{1, n}$ is **positive**, indicating an increase in the energy of the state $\lvert 1, n \rangle$.
- This shift results from the qubit in $\lvert 1 \rangle$ virtually **emitting** a photon into the resonator.

---

## 5. State-Dependent Energy Shifts

**Summary of Results:**

- **For $\lvert 0, n \rangle$:**

  $$
  \Delta E_{0, n} = -\frac{ \hbar g^2 n }{ \Delta }
  $$

- **For $\lvert 1, n \rangle$:**

  $$
  \Delta E_{1, n} = +\frac{ \hbar g^2 ( n + 1 ) }{ \Delta }
  $$

**Key Observations:**

- The energy shifts have **opposite signs** due to the different virtual processes affecting each state.
- The magnitude of the shifts depends on the photon number $n$ and inversely on the detuning $\Delta$.

**Physical Interpretation:**

- **Second-Order Perturbation Effects:** The energy shifts result from second-order perturbation corrections that account for the influence of virtual transitions on the energy levels.
- **State Dependence:** The qubit's ground and excited states experience opposite energy corrections because the virtual absorption and emission processes contribute differently to the energy shifts.

---

## 6. Conclusion

The energy shift due to virtual photon coupling is **state-dependent** because:

- **Second-Order Perturbation Theory:** Virtual transitions contribute to the energy levels in second order, even though they do not conserve energy instantaneously.
- **Opposite Energy Corrections:** The qubit's ground and excited states receive energy shifts of equal magnitude but opposite sign due to these virtual processes.
- **Dependence on Detuning $\Delta$:** The energy shifts are inversely proportional to the detuning between the qubit and resonator frequencies, highlighting the role of the energy difference in the virtual processes.

**Final Result:**

- The energy shift for the qubit in the ground state $\lvert 0 \rangle$ is:

  $$
  \Delta E_{0, n} = -\frac{ \hbar g^2 n }{ \omega_q - \omega_r }
  $$

- The energy shift for the qubit in the excited state $\lvert 1 \rangle$ is:

  $$
  \Delta E_{1, n} = +\frac{ \hbar g^2 ( n + 1 ) }{ \omega_q - \omega_r }
  $$

These energy shifts demonstrate that the qubit's energy levels are affected differently depending on its state due to virtual photon exchange with the resonator, leading to state-dependent energy corrections in the coupled system.
