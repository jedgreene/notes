# Equivalence of First and Second Quantization (Exact Two-Level Example)

Below is a **concrete, fully worked-out example** of a single two-level system—one ground state $\lvert g\rangle$ and one excited state $\lvert e\rangle$—treated in both **first quantization** and **second quantization**.

We will **not** use perturbation theory; instead, we will solve the system **exactly**. In the first-quantized picture, we never speak of “transitions,” we just see how the amplitudes $c_g(t)$ and $c_e(t)$ evolve. In the second-quantized picture, we often **interpret** the coupling term as “annihilating a particle in $g$ and creating it in $e$” (and vice versa)—that is the so-called “transition” language. At the end of the day, **both** descriptions give the **same equations** and **same solutions**, illustrating explicitly that the “transition” view is just a different notation for the same underlying physics.

---

## 1. The Model

Consider a simple Hamiltonian for a two-level system:

$$
\hat{H} \;=\; E_g\,\lvert g\rangle\langle g\rvert 
\;+\;
E_e\,\lvert e\rangle\langle e\rvert
\;+\;
V\bigl(\lvert g\rangle\langle e\rvert \;+\;\lvert e\rangle\langle g\rvert\bigr),
$$

where:

- $E_g$ is the ground-state energy,  
- $E_e$ is the excited-state energy,  
- $V$ is a **real** coupling constant that mixes $\lvert g\rangle$ and $\lvert e\rangle$.

For simplicity, **shift** the zero of energy so that $E_g = 0$. Then

$$
\hat{H}
\;=\;
0 \cdot \lvert g\rangle\langle g\rvert 
\;+\;
\Delta\,\lvert e\rangle\langle e\rvert
\;+\;
V\bigl(\lvert g\rangle\langle e\rvert + \lvert e\rangle\langle g\rvert\bigr),
$$

where $\Delta \equiv E_e - E_g$. In the basis $\{\lvert g\rangle, \lvert e\rangle\}$, the matrix form is:

$$
\hat{H} 
\;=\;
\begin{pmatrix}
0 & V \\
V & \Delta
\end{pmatrix}.
$$

We assume the system is **initially** in the ground state at $t=0$:

$$
\lvert \psi(0)\rangle \;=\; \lvert g\rangle.
$$

---

## 2. First Quantization: No Transition Language

### 2.1 Writing the State and the Schrödinger Equation

In first quantization, the general state is

$$
\lvert \psi(t)\rangle \;=\; c_g(t)\,\lvert g\rangle \;+\; c_e(t)\,\lvert e\rangle,
$$

where $c_g(t)$ and $c_e(t)$ are complex amplitudes with $\lvert c_g\rvert^2 + \lvert c_e\rvert^2 = 1$. The time-dependent Schrödinger equation

$$
i\,\frac{d}{dt}\lvert \psi(t)\rangle 
\;=\;
\hat{H}\,\lvert \psi(t)\rangle
$$

yields (setting $\hbar=1$ for simplicity):

$$
\begin{cases}
i\,\dot{c}_g(t) \;=\; V\,c_e(t),\\
i\,\dot{c}_e(t) \;=\; V\,c_g(t) + \Delta\,c_e(t).
\end{cases}
$$

with initial condition

$$
c_g(0)=1, \quad c_e(0)=0.
$$

### 2.2 Solving the Equations

One can solve this system exactly by diagonalizing the $2\times2$ Hamiltonian matrix or by directly solving the coupled ODEs. The result is the well-known “Rabi solution” form. Regardless of method:

- The amplitude $c_g(t)$ can decrease from 1,
- The amplitude $c_e(t)$ can increase from 0,

reflecting a **flow of amplitude**. **No** mention of “the system transitions” is strictly needed; we just have a single wavefunction in a 2D Hilbert space evolving in time due to the off-diagonal coupling $V$.

---

## 3. Second Quantization: Transition Language

### 3.1 Fock States and Operators

In second quantization, define:

- $a_g^\dagger$ as the creation operator for mode $g$,  
- $a_e^\dagger$ as the creation operator for mode $e$,  
- $\lvert 0\rangle$ as the vacuum (no particles).

Since there is **one** particle in total, the relevant subspace is spanned by

$$
\lvert 1_g, 0_e\rangle \;\equiv\; a_g^\dagger \lvert 0\rangle,
\quad
\lvert 0_g, 1_e\rangle \;\equiv\; a_e^\dagger \lvert 0\rangle.
$$

A general one-particle state is

$$
\lvert \Psi(t)\rangle 
\;=\;
\alpha_g(t)\,\lvert 1_g, 0_e\rangle
\;+\;
\alpha_e(t)\,\lvert 0_g, 1_e\rangle,
$$

with $\lvert \alpha_g\rvert^2 + \lvert \alpha_e\rvert^2 = 1$.

### 3.2 The Same Hamiltonian

The Hamiltonian becomes:

$$
\hat{H} 
\;=\;
0 \cdot (a_g^\dagger a_g)
\;+\;
\Delta\,(a_e^\dagger a_e)
\;+\;
V\bigl(a_g^\dagger a_e + a_e^\dagger a_g\bigr).
$$

### 3.3 Equations of Motion

The Schrödinger equation

$$
i\,\frac{d}{dt}\lvert \Psi(t)\rangle 
\;=\;
\hat{H}\,\lvert \Psi(t)\rangle
$$

gives exactly the same coupled ODEs:

$$
\begin{cases}
i\,\dot{\alpha}_g(t) = V\,\alpha_e(t),\\
i\,\dot{\alpha}_e(t) = V\,\alpha_g(t) + \Delta\,\alpha_e(t),
\end{cases}
$$

with initial condition

$$
\alpha_g(0)=1, \quad \alpha_e(0)=0.
$$

**Identical** to the first-quantized equations for $\{c_g, c_e\}$.

---

## 4. Amplitude Flow vs. “Transition”

- In **first quantization**, we say: “the coupling $V$ drives amplitude transfer from $\lvert g\rangle$ to $\lvert e\rangle$.”
- In **second quantization**, the operator $a_e^\dagger a_g$ can be interpreted as “annihilating a particle in $g$ and creating one in $e$,” i.e., a “transition.”

But **mathematically**, these describe the **same** flow of complex amplitude in a 2D space. There is no difference in predictions, probabilities, or physical outcomes.

---

## 5. Side-by-Side Equivalence

$$
\begin{array}{|c|c|c|}
\hline
\textbf{Aspect} & \textbf{First Quantization} & \textbf{Second Quantization} \\
\hline
\textbf{State} & 
\lvert \psi(t) \rangle = c_g(t)\, \lvert g \rangle + c_e(t)\, \lvert e \rangle & 
\lvert \Psi(t) \rangle = \alpha_g(t)\, \lvert 1_g, 0_e \rangle + \alpha_e(t)\, \lvert 0_g, 1_e \rangle \\
\hline
\textbf{Hamiltonian} & 
\hat{H} = 0\, \lvert g \rangle \langle g \rvert + \Delta\, \lvert e \rangle \langle e \rvert + V\, \big( \lvert g \rangle \langle e \rvert + \lvert e \rangle \langle g \rvert \big) & 
\hat{H} = 0\,(a_g^\dagger a_g) + \Delta\,(a_e^\dagger a_e) + V\,(a_g^\dagger a_e + a_e^\dagger a_g) \\
\hline
\textbf{Time Evolution} & 
i\, \dot{c}_g(t) = V\, c_e(t), \quad i\, \dot{c}_e(t) = V\, c_g(t) + \Delta\, c_e(t) & 
i\, \dot{\alpha}_g(t) = V\, \alpha_e(t), \quad i\, \dot{\alpha}_e(t) = V\, \alpha_g(t) + \Delta\, \alpha_e(t) \\
\hline
\textbf{Initial Condition} & 
c_g(0) = 1, \; c_e(0) = 0 & 
\alpha_g(0) = 1, \; \alpha_e(0) = 0 \\
\hline
\textbf{Interpretation} & 
\text{Amplitude in } \lvert g \rangle \text{ flows to } \lvert e \rangle \text{ via } V & 
a_e^\dagger a_g \text{ "annihilates" in } g \text{ and "creates" in } e \\
\hline
\textbf{Result} & 
\text{Identical final amplitudes: } \{ c_g(t), c_e(t) \} & 
\text{Identical final amplitudes: } \{ \alpha_g(t), \alpha_e(t) \} \\
\hline
\end{array}
$$

Thus, the second-quantized “transition” language is simply an **operator-based notation** for the **same** phenomenon of amplitude flow that appears in first quantization.

---

## 6. Conclusion

**Both** first and second quantization describe the **same** two-level physics:

1. **First Quantization**: A single wavefunction in a 2D Hilbert space, evolving under an off-diagonal Hamiltonian term $V$.  
2. **Second Quantization**: One particle in two possible modes, with creation/annihilation operators that drive “transitions.”

They yield the **exact same** time-dependent solutions. The “transition” language is convenient but **not** mandatory. It is purely an alternative representation of the same amplitude flow seen in the first-quantized picture.
