### Full Derivation of First- and Second-Order Energy and Wavefunction Corrections

We start with the perturbed Hamiltonian $ H = H_0 + \lambda V $ and expand the energy $ E_n $ and wavefunction $ \psi_n $ as power series in $ \lambda $:
$$
E_n = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots
$$
$$
\psi_n = \psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots
$$

### Substitute into the Schrödinger Equation:
$$
(H_0 + \lambda V)(\psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots) = (E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots)(\psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots).
$$

### Collect Terms by Order of $ \lambda $:

#### 1. **Order $ \lambda^0 $**:
$$
H_0 \psi_n^{(0)} = E_n^{(0)} \psi_n^{(0)}.
$$

This is the unperturbed Schrödinger equation.

#### 2. **Order $ \lambda^1 $**:
$$
H_0 \psi_n^{(1)} + V \psi_n^{(0)} = E_n^{(0)} \psi_n^{(1)} + E_n^{(1)} \psi_n^{(0)}.
$$

Rearrange to isolate $ \psi_n^{(1)} $:
$$
(H_0 - E_n^{(0)}) \psi_n^{(1)} = (E_n^{(1)} - V) \psi_n^{(0)}.
$$

### Solving for $ E_n^{(1)} $:
Project onto $ \langle \psi_n^{(0)} | $ to isolate the energy correction:
$$
\langle \psi_n^{(0)} | (H_0 - E_n^{(0)}) \psi_n^{(1)} \rangle = \langle \psi_n^{(0)} | (E_n^{(1)} - V) \psi_n^{(0)} \rangle.
$$

**Simplify Using Hermiticity**:
$$
0 = E_n^{(1)} - \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

**First-Order Energy Correction**:
$$
E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

### Deriving $ \psi_n^{(1)} $:
Return to the first-order equation:
$$
(H_0 - E_n^{(0)}) \psi_n^{(1)} = -V \psi_n^{(0)} + E_n^{(1)} \psi_n^{(0)}.
$$

The term $ E_n^{(1)} \psi_n^{(0)} $ represents the effect of $ V $ that is aligned with $ \psi_n^{(0)} $, contributing only to an energy shift. Since $ \psi_n^{(1)} $ should describe changes orthogonal to $ \psi_n^{(0)} $, we drop this parallel component:
$$
(H_0 - E_n^{(0)}) \psi_n^{(1)} = -V \psi_n^{(0)}.
$$

Expand $ \psi_n^{(1)} $ in terms of the unperturbed eigenstates $ \psi_m^{(0)} $:
$$
\psi_n^{(1)} = \sum_{m \neq n} c_m \psi_m^{(0)}.
$$

Substitute this into the equation:
$$
(H_0 - E_n^{(0)}) \sum_{m \neq n} c_m \psi_m^{(0)} = -V \psi_n^{(0)}.
$$

Acting $ H_0 $ on $ \psi_m^{(0)} $ gives $ E_m^{(0)} \psi_m^{(0)} $, so:
$$
\sum_{m \neq n} c_m (E_m^{(0)} - E_n^{(0)}) \psi_m^{(0)} = -V \psi_n^{(0)}.
$$

Project onto $ \langle \psi_m^{(0)} | $ to isolate $ c_m $:
$$
c_m = \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}}.
$$

**Substitute Back**:
$$
\psi_n^{(1)} = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} \psi_m^{(0)}.
$$

### Second-Order Energy Correction:
Start with:
$$
E_n^{(2)} = \langle \psi_n^{(0)} | V | \psi_n^{(1)} \rangle.
$$

Substitute $ \psi_n^{(1)} $:
$$
E_n^{(2)} = \langle \psi_n^{(0)} | V | \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} \psi_m^{(0)} \rangle.
$$

Expand the inner product:
$$
E_n^{(2)} = \sum_{m \neq n} \frac{\langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}}.
$$

Simplify using the fact that $ \langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle $ is the complex conjugate of $ \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle $:
$$
E_n^{(2)} = \sum_{m \neq n} \frac{|\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle|^2}{E_n^{(0)} - E_m^{(0)}}.
$$

### Deriving $ \psi_n^{(2)} $:
To find the second-order correction $ \psi_n^{(2)} $, we start with the second-order equation:
$$
H_0 \psi_n^{(2)} + V \psi_n^{(1)} = E_n^{(0)} \psi_n^{(2)} + E_n^{(1)} \psi_n^{(1)} + E_n^{(2)} \psi_n^{(0)}.
$$

Rearrange:
$$
(H_0 - E_n^{(0)}) \psi_n^{(2)} = -V \psi_n^{(1)} + E_n^{(1)} \psi_n^{(1)} + E_n^{(2)} \psi_n^{(0)}.
$$

Substitute $ E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle $ and expand $ \psi_n^{(1)} $ using the known expression:
$$
\psi_n^{(2)} = \sum_{k \neq n} \frac{\langle \psi_k^{(0)} | V | \psi_n^{(1)} \rangle}{E_n^{(0)} - E_k^{(0)}} \psi_k^{(0)}.
$$

### Final Results

- **First-Order Energy Correction**:
$$
E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

- **First-Order Wavefunction Correction**:
$$
\psi_n^{(1)} = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} \psi_m^{(0)}.
$$

- **Second-Order Energy Correction**:
$$
E_n^{(2)} = \sum_{m \neq n} \frac{|\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle|^2}{E_n^{(0)} - E_m^{(0)}}.
$$

- **Second-Order Wavefunction Correction**:
$$
\psi_n^{(2)} = \sum_{k \neq n} \frac{\langle \psi_k^{(0)} | V | \psi_n^{(1)} \rangle}{E_n^{(0)} - E_k^{(0)}} \psi_k^{(0)}.
$$
