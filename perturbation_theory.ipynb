# Derivation of First- and Second-Order Corrections in Non-Degenerate Perturbation Theory

In time-independent non-degenerate perturbation theory, we consider a Hamiltonian of the form:

$$
H = H_0 + \lambda V,
$$

where:

- $H_0$ is the unperturbed Hamiltonian.
- $V$ is a small perturbing potential.
- $\lambda$ is a dimensionless parameter indicating the strength of the perturbation.

We aim to find corrections to the energy $E_n$ and the wavefunction $\psi_n$ of the system, expanding them as power series in $\lambda$:

$$
E_n = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots
$$

$$
\psi_n = \psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots
$$

Here, $E_n^{(k)}$ and $\psi_n^{(k)}$ denote the $k$-th order corrections to the energy and wavefunction, respectively.

## Substituting into the Schrödinger Equation

The time-independent Schrödinger equation is:

$$
H \psi_n = E_n \psi_n.
$$

Substituting the expansions of $H$, $E_n$, and $\psi_n$ into this equation, we get:

$$
\left( H_0 + \lambda V \right) \left( \psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots \right) = \left( E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots \right) \left( \psi_n^{(0)} + \lambda \psi_n^{(1)} + \lambda^2 \psi_n^{(2)} + \dots \right).
$$

## Collecting Terms by Order of $\lambda$

By expanding and equating terms of the same order in $\lambda$, we obtain a series of equations.

### Order $\lambda^0$

$$
H_0 \psi_n^{(0)} = E_n^{(0)} \psi_n^{(0)}.
$$

This equation confirms that $\psi_n^{(0)}$ and $E_n^{(0)}$ are the eigenfunctions and eigenvalues of the unperturbed Hamiltonian $H_0$.

### Order $\lambda^1$

Collecting terms of order $\lambda^1$:

$$
H_0 \psi_n^{(1)} + V \psi_n^{(0)} = E_n^{(0)} \psi_n^{(1)} + E_n^{(1)} \psi_n^{(0)}.
$$

Rewriting the equation to group like terms:

$$
\left( H_0 - E_n^{(0)} \right) \psi_n^{(1)} = E_n^{(1)} \psi_n^{(0)} - V \psi_n^{(0)}.
$$

We proceed to solve for the first-order energy correction $E_n^{(1)}$ and the first-order wavefunction correction $\psi_n^{(1)}$.

## Solving for the First-Order Energy Correction $E_n^{(1)}$

To find the first-order energy correction, we take the inner product of both sides with $\langle \psi_n^{(0)} |$:

$$
\langle \psi_n^{(0)} | \left( H_0 - E_n^{(0)} \right) \psi_n^{(1)} \rangle = \langle \psi_n^{(0)} | E_n^{(1)} \psi_n^{(0)} - V \psi_n^{(0)} \rangle.
$$

Since $\left( H_0 - E_n^{(0)} \right)$ acting on $\psi_n^{(0)}$ yields zero and $H_0$ is Hermitian, the left-hand side simplifies to zero:

$$
0 = E_n^{(1)} \langle \psi_n^{(0)} | \psi_n^{(0)} \rangle - \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

Using the normalization $\langle \psi_n^{(0)} | \psi_n^{(0)} \rangle = 1$, we find:

$$
E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

## Deriving the First-Order Wavefunction Correction $\psi_n^{(1)}$

Returning to the first-order equation:

$$
\left( H_0 - E_n^{(0)} \right) \psi_n^{(1)} = E_n^{(1)} \psi_n^{(0)} - V \psi_n^{(0)}.
$$

We apply the orthogonality condition $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$ to ensure that the correction $\psi_n^{(1)}$ is orthogonal to the unperturbed state $\psi_n^{(0)}$. Recognizing that $E_n^{(1)} \psi_n^{(0)}$ does not contribute to $\psi_n^{(1)}$ in the orthogonal subspace, we rewrite the equation as:

$$
\left( H_0 - E_n^{(0)} \right) \psi_n^{(1)} = -\left( V - E_n^{(1)} \right) \psi_n^{(0)}.
$$

We expand $\psi_n^{(1)}$ using the complete set of unperturbed eigenstates $\{ \psi_m^{(0)} \}$:

$$
\psi_n^{(1)} = \sum_{m \neq n} c_m \psi_m^{(0)}.
$$

The term $m = n$ is excluded to satisfy the orthogonality condition $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$.

Substituting the expansion into the equation, we have:

$$
\sum_{m \neq n} c_m \left( E_m^{(0)} - E_n^{(0)} \right) \psi_m^{(0)} = -\left( V - E_n^{(1)} \right) \psi_n^{(0)}.
$$

Taking the inner product with $\langle \psi_m^{(0)} |$ for $m \neq n$:

$$
c_m \left( E_m^{(0)} - E_n^{(0)} \right) = -\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle.
$$

Solving for $c_m$:

$$
c_m = \frac{ \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle }{ E_n^{(0)} - E_m^{(0)} }.
$$

Therefore, the first-order wavefunction correction is:

$$
\psi_n^{(1)} = \sum_{m \neq n} \frac{ \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle }{ E_n^{(0)} - E_m^{(0)} } \psi_m^{(0)}.
$$

## Second-Order Energy Correction $E_n^{(2)}$

From the perturbation expansion, the second-order energy correction is given by:

$$
E_n^{(2)} = \langle \psi_n^{(0)} | V | \psi_n^{(1)} \rangle.
$$

Substituting $\psi_n^{(1)}$ into this expression:

$$
E_n^{(2)} = \sum_{m \neq n} \frac{ \langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle }{ E_n^{(0)} - E_m^{(0)} }.
$$

Recognizing that $V$ is a Hermitian operator, we have:

$$
\langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle = \left( \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle \right)^*.
$$

Therefore, the expression simplifies to:

$$
E_n^{(2)} = \sum_{m \neq n} \frac{ \left| \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle \right|^2 }{ E_n^{(0)} - E_m^{(0)} }.
$$

This result shows that the second-order energy correction arises from the contributions of all other unperturbed states $\psi_m^{(0)}$ coupled to $\psi_n^{(0)}$ via the perturbation $V$.

## Conclusion

Under the framework of time-independent non-degenerate perturbation theory, we have derived:

**First-Order Energy Correction:**

$$
E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle.
$$

**First-Order Wavefunction Correction:**

$$
\psi_n^{(1)} = \sum_{m \neq n} \frac{ \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle }{ E_n^{(0)} - E_m^{(0)} } \psi_m^{(0)}.
$$

**Second-Order Energy Correction:**

$$
E_n^{(2)} = \sum_{m \neq n} \frac{ \left| \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle \right|^2 }{ E_n^{(0)} - E_m^{(0)} }.
$$

These expressions allow us to calculate the approximate energies and states of a quantum system subject to a small perturbation $V$, providing valuable insights in various physical contexts.
