$$

$$
H_m = \frac{\epsilon_0}{2} \int_{V_m} d^3r \left( \vec{\mathbf{E}}^2 + c^2 \vec{\mathbf{B}}^2 \right) = 2 \epsilon_0 V_m \left[ \mathcal{E}_m^{(1)} \right]^2 |\alpha_m(t)|^2

$$

$$

# Coherent States and Classical Oscillations

## Original Quantized Electric Field Operator

The **original form** of the quantized electric field operator for a single mode is:

$$
\hat{\mathbf{E}}(t, \vec{\mathbf{r}}) = i \vec{\epsilon}_m \mathcal{E}_m^{(1)} \left( \hat{a} e^{i(\vec{\mathbf{k}}_m \cdot \vec{\mathbf{r}} - \omega_m t)} - \hat{a}^\dagger e^{-i(\vec{\mathbf{k}}_m \cdot \vec{\mathbf{r}} - \omega_m t)} \right),
$$

where:
- **$i$** is the imaginary unit, contributing to the oscillatory nature of the field.
- **$\vec{\epsilon}_m$** is the **polarization vector** defining the direction of the electric field oscillation.
- **$\mathcal{E}_m^{(1)}$** is a scaling factor related to the field amplitude per photon.
- **$\hat{a}$** is the **annihilation operator**, lowering the photon number by 1.
- **$\hat{a}^\dagger$** is the **creation operator**, increasing the photon number by 1.
- **$\vec{\mathbf{k}}_m$** is the **wave vector** indicating the direction of propagation.
- **$\omega_m$** is the **angular frequency** of the mode.

## Simplified Notation

To make calculations clearer and more focused on the **quantum mechanical behavior**, we often simplify the expression by:
1. **Omitting the polarization vector $\vec{\epsilon}_m$**: It is understood to be present and defining the direction but not explicitly written out.
2. **Dropping the spatial dependence $\vec{\mathbf{k}}_m \cdot \vec{\mathbf{r}}$**: If we are mainly interested in the time-dependent behavior, we can focus on the oscillations over time.
3. **Combining $\mathcal{E}_m^{(1)}$ into a proportionality constant** to avoid clutter.

### Simplified Electric Field Operator

The simplified form becomes:

$$
\hat{\mathbf{E}}(t) \propto i \left( \hat{a} e^{-i\omega_m t} - \hat{a}^\dagger e^{i\omega_m t} \right),
$$

where we understand:
- The **polarization vector $\vec{\epsilon}_m$** is **implicitly present** and applies to the whole expression.
- The term **$i$** still emphasizes the oscillatory nature.
- This form is primarily used to simplify derivations, focusing on the quantum actions of $\hat{a}$ and $\hat{a}^\dagger$.

## Proof: Coherent States Mimicking Classical Oscillations

1. **Starting with the Simplified Electric Field Operator:**
   $$
   \hat{\mathbf{E}}(t) \propto i \left( \hat{a} e^{-i\omega_m t} - \hat{a}^\dagger e^{i\omega_m t} \right).
   $$

2. **Expectation Value in a Coherent State $|\alpha\rangle$:**
   A **coherent state** $|\alpha\rangle$ is an eigenstate of the annihilation operator:
   $$
   \hat{a} |\alpha\rangle = \alpha |\alpha\rangle,
   $$
   where $\alpha$ is a **complex number** representing the **average amplitude and phase** of the field.

3. **Calculating the Expectation Value:**
   The expectation value of the electric field operator in the coherent state $|\alpha\rangle$ is:
   $$
   \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto i \left( \langle \alpha| \hat{a} |\alpha \rangle e^{-i\omega_m t} - \langle \alpha| \hat{a}^\dagger |\alpha \rangle e^{i\omega_m t} \right).
   $$

4. **Using the Eigenvalue Property:**
   Since $\hat{a}|\alpha\rangle = \alpha|\alpha\rangle$ and $\langle \alpha| \hat{a}^\dagger = \langle \alpha| \alpha^*$, we substitute to get:
   $$
   \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto i \left( \alpha e^{-i\omega_m t} - \alpha^* e^{i\omega_m t} \right).
   $$

5. **Expressing $\alpha$ in Terms of Magnitude and Phase:**
   Let $\alpha = |\alpha| e^{i\theta}$, where $|\alpha|$ is the **magnitude** and $\theta$ is the **phase**. Then:
   $$
   \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto i \left( |\alpha| e^{i(\theta - \omega_m t)} - |\alpha| e^{-i(\theta - \omega_m t)} \right).
   $$

6. **Simplify Using Euler's Formula:**
   Recall that $e^{i\phi} - e^{-i\phi} = 2i \sin(\phi)$. Thus:
   $$
   \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto 2 |\alpha| \sin(\theta - \omega_m t).
   $$

7. **Final Expression: Oscillatory Behavior:**
   $$
   \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto |\alpha| \sin(\omega_m t - \theta).
   $$

## Interpretation and Connection to Classical Behavior

1. **Oscillatory Behavior:**
   - The expectation value shows a **sinusoidal oscillation** similar to a classical wave, with an amplitude $|\alpha|$ and a phase shift $\theta$.
   - This demonstrates that the quantized field in a **coherent state** behaves like a classical oscillating field.

2. **Implicit Direction and Polarization:**
   - Even though we used a **simplified form** for easier calculation, the **polarization vector $\vec{\epsilon}_m$** is implicitly present, ensuring that the field oscillates in the correct direction, just as it would classically.

## Conclusion

By starting with the **original form** of the quantized electric field operator and then moving to a **simplified notation**, we can more easily perform calculations that demonstrate how **coherent states** behave like classical oscillating waves. The simplification helps focus on the core quantum mechanical behavior, while it is understood that details like **polarization** are still present. This approach shows that coherent states bridge the gap between classical and quantum descriptions, allowing the field to mimic classical oscillatory behavior.

# Intuition Behind the Proof: Coherent States Mimicking Classical Oscillations

## Classical vs. Quantum Behavior

1. **Classical Oscillating Field:**
   - In classical physics, an electric field can be thought of as a **wave** that oscillates back and forth, characterized by a smooth, continuous function like:
     $$
     \mathbf{E}(t) = \mathcal{E}_0 \sin(\omega t - \theta),
     $$
     where:
     - $\mathcal{E}_0$ is the **amplitude**, determining how strong the oscillation is.
     - $\omega$ is the **angular frequency**, determining how fast it oscillates.
     - $\theta$ is the **phase**, determining where the peaks and troughs occur over time.
   - This wave-like behavior is continuous, and the energy is also continuous (you can have any amount of energy depending on the amplitude $\mathcal{E}_0$).

2. **Quantized Field in Quantum Mechanics:**
   - When we move to quantum mechanics, we **quantize** the electromagnetic field. This means that the field is no longer a smooth, continuous wave but is made up of discrete energy packets called **photons**.
   - The electric field becomes an **operator** that acts on quantum states, and the energy is no longer continuous but comes in multiples of $\hbar \omega$, where $\hbar$ is the reduced Planck constant, and $\omega$ is the frequency.

## What the Proof Shows: Bridging Classical and Quantum

1. **Coherent States as "Classical-Like" States:**
   - In quantum mechanics, **coherent states** $|\alpha\rangle$ are special because they are the closest analog to a classical oscillating wave.
   - The state $|\alpha\rangle$ is defined by a **complex number** $\alpha$, which determines the **average amplitude and phase** of the field, much like $\mathcal{E}_0$ and $\theta$ do in the classical case.

2. **Expectation Value Mimics Classical Oscillation:**
   - The proof calculates the **expectation value** of the electric field operator $\hat{\mathbf{E}}(t)$ when the field is in a coherent state $|\alpha\rangle$.
   - The expectation value $\langle \alpha | \hat{\mathbf{E}}(t) | \alpha \rangle$ is a way of asking, "On average, how does the field behave when it's in this quantum state?"
   - The result of the proof is that this expectation value looks like:
     $$
     \langle \alpha| \hat{\mathbf{E}}(t) |\alpha \rangle \propto |\alpha| \sin(\omega t - \theta),
     $$
     which is **exactly the same form as a classical oscillating wave**.

3. **Intuitive Meaning:**
   - Even though the field is fundamentally **quantized** and made up of discrete photons, when it's in a **coherent state**, it **oscillates** in a way that looks very much like a **classical wave**. 
   - The complex number $\alpha$ acts like the classical amplitude $\mathcal{E}_0$ and phase $\theta$. The **magnitude** $|\alpha|$ determines how strong the oscillation is (similar to the amplitude), and the **argument** $\theta$ determines the phase of the oscillation.
   - This is why coherent states are often called the "most classical" of quantum states. They **minimize quantum uncertainty** and produce behavior that resembles classical oscillations.

## Why This is Significant

1. **Bridging Classical and Quantum:**
   - This result shows how quantum mechanics can **replicate** classical physics under certain conditions. Even though the electromagnetic field is quantized, it can still behave like a classical wave when it's in a coherent state.
   - This is important for understanding how classical concepts, like electromagnetic waves or oscillations, emerge from quantum mechanics. It shows that classical behavior isn't lost but is rather a **special case** of quantum behavior.

2. **Practical Relevance:**
   - **Lasers** are a real-world example of this phenomenon. The light emitted by a laser is in a coherent state, which is why it behaves so much like a classical wave. It has a well-defined amplitude and phase and can propagate over long distances without spreading out like typical light.
   - In **quantum optics** and related fields, coherent states are used to model light that can be treated as almost classical, making calculations simpler and more intuitive.

## Summary

The proof shows that the **expectation value** of the electric field operator in a **coherent state** behaves just like a **classical oscillating wave**. This demonstrates that even though the field is quantized and made up of discrete photons, it can still exhibit classical behavior when in the right state. Coherent states, therefore, act as a **bridge** between the classical and quantum worlds, allowing us to understand how classical oscillations can emerge from the underlying quantum field.


# Connection Between Coherent States and the Uncertainty Principle

## The Heisenberg Uncertainty Principle

The **Heisenberg Uncertainty Principle** states that there is a fundamental limit to how precisely we can know certain pairs of properties (like position and momentum) simultaneously. For the electromagnetic field, the relevant quantities are:

- **Field quadratures**, which are analogous to position and momentum. These quadratures can be thought of as two components of the electric field that vary in time.
  
Mathematically, if we denote two conjugate field quadratures by $\hat{X}$ and $\hat{P}$, the uncertainty principle is expressed as:
$$
\Delta X \cdot \Delta P \geq \frac{\hbar}{2},
$$
where:
- $\Delta X$ is the standard deviation (uncertainty) in the measurement of $\hat{X}$,
- $\Delta P$ is the standard deviation (uncertainty) in the measurement of $\hat{P}$,
- $\hbar$ is the reduced Planck constant.

## Coherent States and Minimum Uncertainty

1. **Coherent States Minimize the Uncertainty:**
   - Coherent states are **special quantum states** that **minimize** the uncertainty between two conjugate variables (like $\hat{X}$ and $\hat{P}$). They satisfy the equality:
     $$
     \Delta X \cdot \Delta P = \frac{\hbar}{2},
     $$
     meaning that they are the **least uncertain** states allowed by quantum mechanics.
   - This makes coherent states different from other quantum states, which might have greater uncertainty in one or both of these variables. For example, in a **number state** (a state with a fixed number of photons), the phase is completely uncertain, leading to larger uncertainties.

2. **Balanced Uncertainty:**
   - In a coherent state, the uncertainties $\Delta X$ and $\Delta P$ are **equal**. Unlike other states where one uncertainty might be squeezed down at the expense of the other (leading to states called **squeezed states**), coherent states keep a perfect balance.
   - This balanced uncertainty allows coherent states to have **well-defined amplitudes and phases**, making them behave much like classical waves.

## Connection to Classical Behavior

1. **Classical-Like Oscillations:**
   - Because coherent states minimize the uncertainty between the field quadratures, the electric field’s oscillations in a coherent state appear **smooth and regular**, just like in classical physics. 
   - This is why, when we take the **expectation value** of the electric field operator in a coherent state, we see a **sinusoidal wave** that resembles a classical oscillation:
     $$
     \langle \alpha | \hat{\mathbf{E}}(t) | \alpha \rangle \propto |\alpha| \sin(\omega t - \theta).
     $$

2. **Predictable Amplitude and Phase:**
   - In classical waves, the amplitude and phase are **well-defined**, allowing for predictable, stable oscillations. Coherent states, by minimizing the uncertainty, achieve something similar in the quantum realm: they maintain a **predictable average amplitude** $|\alpha|$ and **phase** $\theta$.
   - Although quantum mechanics is inherently probabilistic, coherent states provide a way for the field to behave in a **stable, classical-like** manner.

3. **Contrast with Other Quantum States:**
   - For other quantum states, like **number states** (states with a precise number of photons), there is much more uncertainty in the phase. This is why we don’t see classical-like oscillations from these states. Instead, they exhibit more **quantum noise**, making them less predictable and more distinct from classical behavior.
   - Coherent states' ability to **minimize noise** and maintain **balanced uncertainty** allows them to exhibit behavior that is **close to classical**.

## Practical Implications: Lasers and Light Beams

1. **Lasers as Coherent Light Sources:**
   - A **laser** is essentially a source of light in a coherent state. This is why laser light can be so stable and well-behaved over long distances: the **amplitude** and **phase** are predictable, and the light waves oscillate smoothly, much like a classical wave.
   - This predictability and stability are possible because the light emitted by the laser is in a coherent state, which minimizes uncertainty between amplitude and phase.

2. **Classical-Quantum Transition:**
   - Coherent states provide an excellent example of how **classical behavior emerges from quantum mechanics**. Even though photons are inherently quantum particles, when they are in a coherent state, the field they create behaves as if it were a classical electromagnetic wave.
   - This demonstrates a deep connection between **quantum mechanics** and **classical physics**, showing that under certain conditions, quantum systems can mimic classical behavior.

## Summary

The connection between **coherent states** and the **Heisenberg Uncertainty Principle** lies in the fact that coherent states **minimize the uncertainty** between two key variables (field quadratures), achieving the least amount of quantum noise possible. This minimization allows coherent states to maintain **stable, predictable oscillations** of the electric field, much like a classical wave. 

This is why, when we take the expectation value of the electric field operator in a coherent state, we observe behavior that looks very **classical**. The minimization of uncertainty ensures that **amplitude and phase** are well-defined, leading to smooth, regular oscillations similar to classical electromagnetic waves. Coherent states, therefore, act as a **bridge** between the quantum and classical worlds, showing how classical phenomena can emerge from the quantum domain.
